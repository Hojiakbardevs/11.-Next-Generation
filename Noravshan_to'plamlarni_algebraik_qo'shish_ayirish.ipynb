```python
# Define the fuzzy numbers A and B as (central value, left spread, right spread)
A = (5, 1, 2)
B = (3, 0.5, 1.5)

# Addition of A and B
def fuzzy_add(A, B):
    return (A[0] + B[0], A[1] + B[1], A[2] + B[2])

# Subtraction of A and B
def fuzzy_subtract(A, B):
    return (A[0] - B[0], A[1] + B[1], A[2] + B[2])

# Multiplication of A and B
def fuzzy_multiply(A, B):
    return (A[0] * B[0], abs(A[0]) * B[1] + abs(B[0]) * A[1], abs(A[0]) * B[2] + abs(B[0]) * A[2])

# Division of A by B (assuming B's central value is not zero)
def fuzzy_divide(A, B):
    if B[0] == 0:
        return None  # Division by zero not defined
    central = A[0] / B[0]
    left_spread = (abs(A[0]) * B[1] + abs(B[0]) * A[1]) / (B[0] ** 2)
    right_spread = (abs(A[0]) * B[2] + abs(B[0]) * A[2]) / (B[0] ** 2)
    return (central, left_spread, right_spread)

# Perform and print the results of each operation
add_result = fuzzy_add(A, B)
subtract_result = fuzzy_subtract(A, B)
multiply_result = fuzzy_multiply(A, B)
divide_result = fuzzy_divide(A, B)

print("Addition:", add_result)
print("Subtraction:", subtract_result)
print("Multiplication:", multiply_result)
print("Division:", divide_result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Noravshan sonlar uchun L-R funktsiyalari
class LR_Fuzzy:
    def __init__(self, a, aL, aR):
        self.a = a     # Markaziy qiymat
        self.aL = aL   # Chap kenglik
        self.aR = aR   # O'ng kenglik

    # Qo‘shish
    def add(self, other):
        a = self.a + other.a
        aL = self.aL + other.aL
        aR = self.aR + other.aR
        return LR_Fuzzy(a, aL, aR)

    # Ayirish
    def subtract(self, other):
        a = self.a - other.a
        aL = self.aL + other.aL
        aR = self.aR + other.aR
        return LR_Fuzzy(a, aL, aR)

    # Ko‘paytirish
    def multiply(self, other):
        a = self.a * other.a
        aL = abs(self.a) * other.aL + abs(other.a) * self.aL
        aR = abs(self.a) * other.aR + abs(other.a) * self.aR
        return LR_Fuzzy(a, aL, aR)

    # Bo‘lish (faqat other.a ≠ 0 bo'lsa)
    def divide(self, other):
        if other.a != 0:
            a = self.a / other.a
            aL = (abs(self.a) * other.aL + abs(other.a) * self.aL) / (other.a ** 2)
            aR = (abs(self.a) * other.aR + abs(other.a) * self.aR) / (other.a ** 2)
            return LR_Fuzzy(a, aL, aR)
        else:
            raise ValueError("Bo'linuvchi 0 ga teng bo'lishi mumkin emas.")

# Grafik chizish funktsiyasi
def plot_fuzzy_number(fuzzy, color, label):
    x_left = np.linspace(fuzzy.a - fuzzy.aL, fuzzy.a, 100)
    y_left = np.linspace(0, 1, 100)
    x_right = np.linspace(fuzzy.a, fuzzy.a + fuzzy.aR, 100)
    y_right = np.linspace(1, 0, 100)

    plt.plot(x_left, y_left, color=color, linestyle='-', label=label)
    plt.plot(x_right, y_right, color=color, linestyle='-')
    plt.fill_between(np.concatenate([x_left, x_right]), np.concatenate([y_left, y_right]), alpha=0.2, color=color)

# Ikkita L-R noravshan sonlar
A = LR_Fuzzy(170, 10, 15)
B = LR_Fuzzy(160, 8, 12)

# Amallarni bajarish
sum_result = A.add(B)
sub_result = A.subtract(B)
mul_result = A.multiply(B)
div_result = A.divide(B)

# Grafik chizish
plt.figure(figsize=(12, 8))
plot_fuzzy_number(A, color='blue', label='A - Odam bo\'yi')
plot_fuzzy_number(B, color='green', label='B - Odam bo\'yi')

plot_fuzzy_number(sum_result, color='orange', label='A + B (Qo‘shish)')
plot_fuzzy_number(sub_result, color='red', label='A - B (Ayirish)')
plot_fuzzy_number(mul_result, color='purple', label='A * B (Ko‘paytirish)')
plot_fuzzy_number(div_result, color='brown', label='A / B (Bo‘lish)')

# Grafikni bezash
plt.xlabel("Qiymatlar")
plt.ylabel("A'zolik darajasi")
plt.title("L-R Noravshan Sonlar bilan Algebraik Amallar")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class LR_Fuzzy:
    def __init__(self, a, aL, aR):
        self.a = float(a)    # Central value
        self.aL = float(aL)  # Left width
        self.aR = float(aR)  # Right width

    def __str__(self):
        return f"({self.a - self.aL}, {self.a}, {self.a + self.aR})"

    # Operator overloading for addition (+)
    def __add__(self, other):
        if isinstance(other, (int, float)):
            # Adding a crisp number
            return LR_Fuzzy(self.a + other, self.aL, self.aR)
        else:
            # Adding another fuzzy number
            return LR_Fuzzy(self.a + other.a, self.aL + other.aL, self.aR + other.aR)

    # Operator overloading for subtraction (-)
    def __sub__(self, other):
        if isinstance(other, (int, float)):
            # Subtracting a crisp number
            return LR_Fuzzy(self.a - other, self.aL, self.aR)
        else:
            # Subtracting another fuzzy number
            return LR_Fuzzy(self.a - other.a, self.aL + other.aL, self.aR + other.aR)

    # Operator overloading for multiplication (*)
    def __mul__(self, other):
        if isinstance(other, (int, float)):
            # Multiplying by a crisp number
            if other >= 0:
                return LR_Fuzzy(self.a * other, self.aL * other, self.aR * other)
            else:
                return LR_Fuzzy(self.a * other, self.aR * abs(other), self.aL * abs(other))
        else:
            # Multiplying by another fuzzy number
            a = self.a * other.a
            aL = abs(self.a) * other.aL + abs(other.a) * self.aL
            aR = abs(self.a) * other.aR + abs(other.a) * self.aR
            return LR_Fuzzy(a, aL, aR)

    # Operator overloading for division (/)
    def __truediv__(self, other):
        if isinstance(other, (int, float)):
            if other != 0:
                # Dividing by a crisp number
                if other > 0:
                    return LR_Fuzzy(self.a / other, self.aL / other, self.aR / other)
                else:
                    return LR_Fuzzy(self.a / other, self.aR / abs(other), self.aL / abs(other))
            else:
                raise ValueError("Division by zero")
        else:
            # Dividing by another fuzzy number
            if other.a != 0:
                a = self.a / other.a
                aL = (abs(self.a) * other.aL + abs(other.a) * self.aL) / (other.a ** 2)
                aR = (abs(self.a) * other.aR + abs(other.a) * self.aR) / (other.a ** 2)
                return LR_Fuzzy(a, aL, aR)
            else:
                raise ValueError("Division by zero")

def plot_fuzzy_number(fuzzy, color, label, ax):
    x_left = np.linspace(fuzzy.a - fuzzy.aL, fuzzy.a, 100)
    y_left = np.linspace(0, 1, 100)
    x_right = np.linspace(fuzzy.a, fuzzy.a + fuzzy.aR, 100)
    y_right = np.linspace(1, 0, 100)

    ax.plot(x_left, y_left, color=color, linestyle='-', label=label)
    ax.plot(x_right, y_right, color=color, linestyle='-')
    ax.fill_between(np.concatenate([x_left, x_right]),
                   np.concatenate([y_left, y_right]),
                   alpha=0.2, color=color)

def print_operation_details(A, B, operation, result):
    print(f"\n{operation} Operation Details:")
    print(f"A = {A}")
    print(f"B = {B}")
    print(f"Result = {result}")
    print(f"Support interval: [{result.a - result.aL}, {result.a + result.aR}]")
    print(f"Core value: {result.a}")
    print(f"Left spread: {result.aL}")
    print(f"Right spread: {result.aR}")

# Create example fuzzy numbers
A = LR_Fuzzy(170, 10, 15)  # First fuzzy number
B = LR_Fuzzy(160, 8, 12)   # Second fuzzy number

# Now you can use operators directly
C_result = A + B      # Addition
D_result = A - B      # Subtraction
E_result = A * B      # Multiplication
F_result = A / B      # Division

# You can also mix fuzzy numbers with regular numbers
G_result = A + 5      # Adding a crisp number
H_result = A * 2      # Multiplying by a crisp number

# Print results
print("\nFuzzy Number Operations with Operators:")
print(f"A + B = {C_result}")
print(f"A - B = {D_result}")
print(f"A * B = {E_result}")
print(f"A / B = {F_result}")
print(f"A + 5 = {G_result}")
print(f"A * 2 = {H_result}")

# Create visualization
fig, axs = plt.subplots(3, 2, figsize=(15, 18))
plt.suptitle("Extended Fuzzy Number Operations", fontsize=16)

# Plot original numbers
axs[0, 0].set_title('Original Fuzzy Numbers')
plot_fuzzy_number(A, 'blue', 'A', axs[0, 0])
plot_fuzzy_number(B, 'green', 'B', axs[0, 0])
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot addition
axs[0, 1].set_title('Addition (A + B)')
plot_fuzzy_number(C_result, 'orange', 'A + B', axs[0, 1])
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot subtraction
axs[1, 0].set_title('Subtraction (A - B)')
plot_fuzzy_number(D_result, 'red', 'A - B', axs[1, 0])
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot multiplication
axs[1, 1].set_title('Multiplication (A * B)')
plot_fuzzy_number(E_result, 'purple', 'A * B', axs[1, 1])
axs[1, 1].legend()
axs[1, 1].grid(True)

# Plot division and crisp number operation
axs[2, 0].set_title('Division (A / B)')
plot_fuzzy_number(F_result, 'brown', 'A / B', axs[2, 0])
axs[2, 0].legend()
axs[2, 0].grid(True)

# Plot crisp number operations
axs[2, 1].set_title('Crisp Number Operations')
plot_fuzzy_number(G_result, 'cyan', 'A + 5', axs[2, 1])
plot_fuzzy_number(H_result, 'magenta', 'A * 2', axs[2, 1])
axs[2, 1].legend()
axs[2, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class LR_Fuzzy:
    def __init__(self, a, aL, aR):
        self.a = float(a)    # Central value
        self.aL = float(aL)  # Left width
        self.aR = float(aR)  # Right width

    def __str__(self):
        return f"({self.a - self.aL}, {self.a}, {self.a + self.aR})"

    def add(self, other):
        a = self.a + other.a
        aL = self.aL + other.aL
        aR = self.aR + other.aR
        return LR_Fuzzy(a, aL, aR)

    def subtract(self, other):
        a = self.a - other.a
        aL = self.aL + other.aL
        aR = self.aR + other.aR
        return LR_Fuzzy(a, aL, aR)

    def multiply(self, other):
        a = self.a * other.a
        aL = abs(self.a) * other.aL + abs(other.a) * self.aL
        aR = abs(self.a) * other.aR + abs(other.a) * self.aR
        return LR_Fuzzy(a, aL, aR)

    def divide(self, other):
        if other.a != 0:
            a = self.a / other.a
            aL = (abs(self.a) * other.aL + abs(other.a) * self.aL) / (other.a ** 2)
            aR = (abs(self.a) * other.aR + abs(other.a) * self.aR) / (other.a ** 2)
            return LR_Fuzzy(a, aL, aR)
        else:
            raise ValueError("Division by zero is not allowed")

def plot_fuzzy_number(fuzzy, color, label, ax):
    x_left = np.linspace(fuzzy.a - fuzzy.aL, fuzzy.a, 100)
    y_left = np.linspace(0, 1, 100)
    x_right = np.linspace(fuzzy.a, fuzzy.a + fuzzy.aR, 100)
    y_right = np.linspace(1, 0, 100)

    ax.plot(x_left, y_left, color=color, linestyle='-', label=label)
    ax.plot(x_right, y_right, color=color, linestyle='-')
    ax.fill_between(np.concatenate([x_left, x_right]),
                   np.concatenate([y_left, y_right]),
                   alpha=0.2, color=color)

def print_operation_details(A, B, operation, result):
    print(f"\n{operation} Operation Details:")
    print(f"A = {A}")
    print(f"B = {B}")
    print(f"Result = {result}")
    print(f"Support interval: [{result.a - result.aL}, {result.a + result.aR}]")
    print(f"Core value: {result.a}")
    print(f"Left spread: {result.aL}")
    print(f"Right spread: {result.aR}")

# Create example fuzzy numbers
A = LR_Fuzzy(170, 10, 15)  # First fuzzy number
B = LR_Fuzzy(160, 8, 12)   # Second fuzzy number

# Perform operations
sum_result = A.add(B)
sub_result = A.subtract(B)
mul_result = A.multiply(B)
div_result = A.divide(B)

# Print detailed results for each operation
print("\nFuzzy Number Details:")
print(f"A: Central value = {A.a}, Left spread = {A.aL}, Right spread = {A.aR}")
print(f"B: Central value = {B.a}, Left spread = {B.aL}, Right spread = {B.aR}")

print_operation_details(A, B, "Addition", sum_result)
print_operation_details(A, B, "Subtraction", sub_result)
print_operation_details(A, B, "Multiplication", mul_result)
print_operation_details(A, B, "Division", div_result)

# Create visualization
fig, axs = plt.subplots(2, 2, figsize=(15, 12))
plt.suptitle("Fuzzy Number Operations Visualization", fontsize=16)

# Plot original numbers
axs[0, 0].set_title('Original Fuzzy Numbers')
plot_fuzzy_number(A, 'blue', 'A', axs[0, 0])
plot_fuzzy_number(B, 'green', 'B', axs[0, 0])
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot addition
axs[0, 1].set_title('Addition (A + B)')
plot_fuzzy_number(sum_result, 'orange', 'A + B', axs[0, 1])
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot subtraction
axs[1, 0].set_title('Subtraction (A - B)')
plot_fuzzy_number(sub_result, 'red', 'A - B', axs[1, 0])
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot multiplication and division
axs[1, 1].set_title('Multiplication (A * B) and Division (A / B)')
plot_fuzzy_number(mul_result, 'purple', 'A * B', axs[1, 1])
plot_fuzzy_number(div_result, 'brown', 'A / B', axs[1, 1])
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
